```python
from sklearn.model_selection import train_test_split
b_train,b_test,x_train,x_test=train_test_split(b,x,test_size=0.30,shuffle=True,random_state=40)
print("b shape is ",b_train.shape)
print("b shape is ",b_test.shape)
print("x shape is ",x_train.shape)
print("x shape is ",x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(b_train,x_train)
```

```python
plt.plot(b_train,x_train,'mo')
```

```python
plt.plot(b_test,x_test,'bo')
```

```python
x_pre=lr.predict(b_test)
plt.plot(b_test,x_test,'bo')
plt.plot(b_test,x_test,'k-')
```

```python
lr.score(b_test,x_test)
```
